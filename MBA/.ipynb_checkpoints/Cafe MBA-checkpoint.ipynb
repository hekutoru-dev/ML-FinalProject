# Cafe Transaction Data

* Import dependencies.

```python
import pandas as pd
import numpy as np
```

```python
cafe_data = pd.read_csv('CafeDataMBA.csv', encoding='latin-1')
cafe_data
```

```python
cafe_data.columns
```

* Rename Columns.

```python
cafe_data = cafe_data.rename(columns={'Bill Number ':'BillNumber',
                                      'Item Desc':'Description',
                                      'day':'Day',
                                      'Time Bucket':'TimeBucket'})
```

```python
cafe_data.columns
```

* Drop unnecessary information (not columns).

Drop information, apparently columns, which are just tip information of the Bucket Time to understand what the TimeBucket column is.

```python
cafe_data = cafe_data.drop(['Unnamed: 14', 'start', 'Bucket'], axis=1)
cafe_data.columns
```

* Check for possible missing values.

```python
cafe_data.count()
```

```python
cafe_data['Description'].value_counts().head(10)
```

```python
cafe_data['Description'].unique()
```

* Rename values in the Description item column that contains strange data.

```python
cafe_data = cafe_data.replace({'Description': {'WHITE SANGRIA (GLS)áááá áááááá' : 'WHITE SANGRIA (GLS)',
                                               'WHITE SANGRIA (CARAFE)áááááááá' : 'WHITE SANGRIA (CARAFE)',
                                               'RED SANGRIA (GLS)á áááááááááá ' : 'RED SANGRIA (GLS)',
                                               'RED SANGRIA (CARAFE) áááááááá ' : 'RED SANGRIA (CARAFE)',
                                               'C C C ROMA TOMATO SALAD       ' : 'ROMA TOMATO SALAD',
                                               'THAT CHOCOLATE THING......... ' : 'THAT CHOCOLATE THING'}})
```

```python
cafe_data['Description'].unique()
```
