# PROFITABILITY ANALYSIS

```python
#Create Profits Columns
cafe_data['Profits'] = (cafe_data['Rate'] - cafe_data['Cost'])
cafe_data.head()
```

```python
#Visualize categories profitability (high to low)
categories_2 = cafe_data.groupby(['Category2']).sum().reset_index()
categories_2 = categories_2.drop(['Month', 'Year'], axis = 1)
#categories_2.div(1000000)
categories_2.sort_values('Rate', ascending=False)
```

```python
type(categories_2)
```

```python
# Transform values into a more readable view
categories_2['Rate_000s'] = categories_2['Rate']/(1000)
categories_2['Tax_000s'] = categories_2['Tax']/(1000)
categories_2['Discount_000s'] = categories_2['Discount']/(1000)
categories_2['Total_000s'] = categories_2['Total']/(1000)
categories_2['Cost_000s'] = categories_2['Cost']/(1000)
categories_2['Profits_000s'] = categories_2['Profits']/(1000)
categories2_K = categories_2[['Category2', 'Quantity', 'Rate_000s', 'Tax_000s', 'Discount_000s', 'Total_000s', 'Cost_000s', 'Profits_000s' ]]
categories2_K.sort_values('Profits_000s', ascending=False)
```

```python
# Perform an analysis by time bucket
time_bkts = cafe_data.groupby(['TimeBucket'])
profits_by_bkt = time_bkts['Profits'].sum()/1000
profits_by_bkt
```

```python
grp_products = cafe_data.groupby(['Category2', 'TimeBucket']).sum().reset_index()
grp_products = grp_products.drop(['Month', 'Year', 'Rate', 'Tax', 'Discount'], axis=1)
grp_products.sort_values('Total', ascending=False)
print(grp_products)
```

```python
tobacco_analysis = cafe_data.loc[cafe_data['Category2'] == 'TOBACCO']
tobacco_time = tobacco_analysis.groupby(['TimeBucket']).sum().reset_index()
tobacco_time.sort_values('Profits', ascending=False)
```

```python
shakes_analysis = cafe_data.loc[cafe_data['Category2'] == 'SHAKES']
shakes_time = shakes_analysis.groupby(['TimeBucket']).sum().reset_index()
shakes_time.sort_values('Profits', ascending=False)
```

```python
beer_analysis = cafe_data.loc[cafe_data['Category2'] == 'BEER']
beer_time = beer_analysis.groupby(['TimeBucket']).sum().reset_index()
beer_time.sort_values('Profits', ascending=False)
```
